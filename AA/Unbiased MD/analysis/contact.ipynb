```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

# 103
def y_formatter(x, pos):
    return f'{int(x / 1000)}'  # 1000

# Function to calculate mean and standard error
def calculate_mean_and_error(ydata_list):
    # Get the last 5000 data points from each dataset
    last_5000_data = [ydata[-5000:] for ydata in ydata_list]
    # 
    means = [np.mean(data) for data in last_5000_data]
    errors = [np.std(data) for data in last_5000_data]  # 
    return means, errors

#  Nature 
plt.rcParams.update({
    'font.family': 'Arial',  # 
    'axes.titlesize': 36,  # 
    'axes.labelsize': 36,  # 
    'xtick.labelsize': 36,  # X
    'ytick.labelsize': 36,  # Y
    'legend.fontsize': 36,  # 
    'figure.figsize': (8, 6),  # 
    'axes.linewidth': 4,  # 
    'xtick.major.size': 10,  # x 
    'ytick.major.size': 10,  # y 
    'xtick.minor.size': 5,  # x 
    'ytick.minor.size': 5,  # y 
    'xtick.major.width': 4,  # x 
    'ytick.major.width': 4,  # y 
    'xtick.minor.width': 3,  # x 
    'ytick.minor.width': 3,  # y 
    'axes.grid': False,  # 
})

# 
mean1, error1 = calculate_mean_and_error([ydata1, ydata2, ydata3])
mean2, error2 = calculate_mean_and_error([ydata4, ydata5, ydata6])
mean3, error3 = calculate_mean_and_error([ydata7, ydata8, ydata9])

means = [np.mean(mean1), np.mean(mean2), np.mean(mean3)]
errors = [np.std(mean1,ddof=1)/ np.sqrt(len(mean1)), np.std(mean2,ddof=1)/ np.sqrt(len(mean2)), np.std(mean3,ddof=1)/ np.sqrt(len(mean3))]

# 
fig, ax = plt.subplots(figsize=(8.5, 7))
x = np.arange(2)  # 

# 
bars1 = ax.bar(x[0], means[0], yerr=errors[0], capsize=5, color='tab:red',
               error_kw={'elinewidth': 4, 'capthick': 4}, edgecolor='black', linewidth=4)

# 
bars2 = ax.bar(x[1], means[1], yerr=errors[1], capsize=5, color='tab:green',
               error_kw={'elinewidth': 4, 'capthick': 4}, edgecolor='black', linewidth=4)
bars3 = ax.bar(x[1], means[2], yerr=errors[2], capsize=5, color='tab:purple',
               error_kw={'elinewidth': 4, 'capthick': 4}, edgecolor='black', linewidth=4, bottom=means[1])

# 
ax.text(x[0], means[0] / 2, 'PROTEIN', ha='center', va='center', fontsize=25, color='white', weight='bold')
ax.text(x[1], means[1] / 2, 'PROTEIN', ha='center', va='center', fontsize=25, color='white', weight='bold')
ax.text(x[1], (means[1] + means[2])* 9/10, 'DNA', ha='center', va='center', fontsize=25, color='white', weight='bold')

# 
ax.set_ylabel(r'Contact number ($\times 10^3$)')
ax.set_title('')

# x 
ax.set_xticks(x)
ax.set_xticklabels(['US', 'BS'])

# y 
ax.tick_params(axis='y', which='both', direction='in')
# x 
ax.tick_params(axis='x', which='both', direction='in')

ax.yaxis.set_major_formatter(ticker.FuncFormatter(y_formatter))
#  y 5000
ax.yaxis.set_major_locator(ticker.MultipleLocator(5000))
# 
ax.legend(loc='upper right', frameon=False)
ax.set_ylim(0,15000)

# 
plt.subplots_adjust(left=0.2, right=0.9, top=0.9, bottom=0.15)  # 

#  bbox_inches='tight'
plt.savefig("Mean_Values_contact.png", dpi=300)
plt.show()
```
