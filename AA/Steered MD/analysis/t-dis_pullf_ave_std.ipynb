```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
from matplotlib.ticker import ScalarFormatter

# 读取数据
data1x = np.loadtxt('/data/jychen/MD_projects/SMC/All_Atoms/SteeredMD/2SMD_1protonataing/1/pull_pullx.xvg', comments=['@', '#'])
data2x = np.loadtxt('/data/jychen/MD_projects/SMC/All_Atoms/SteeredMD/2SMD_1protonataing/2/pull_pullx.xvg', comments=['@', '#'])
data3x = np.loadtxt('/data/jychen/MD_projects/SMC/All_Atoms/SteeredMD/2SMD_1protonataing/3/pull_pullx.xvg', comments=['@', '#'])
data1y = np.loadtxt('/data/jychen/MD_projects/SMC/All_Atoms/SteeredMD/2SMD_1protonataing/1/pull_pullf.xvg', comments=['@', '#'])
data2y = np.loadtxt('/data/jychen/MD_projects/SMC/All_Atoms/SteeredMD/2SMD_1protonataing/2/pull_pullf.xvg', comments=['@', '#'])
data3y = np.loadtxt('/data/jychen/MD_projects/SMC/All_Atoms/SteeredMD/2SMD_1protonataing/3/pull_pullf.xvg', comments=['@', '#'])
data4x = np.loadtxt('/data/jychen/MD_projects/SMC/All_Atoms/SteeredMD/2SMD_1protonataing+DNA/1/pull_pullx.xvg', comments=['@', '#'])
data5x = np.loadtxt('/data/jychen/MD_projects/SMC/All_Atoms/SteeredMD/2SMD_1protonataing+DNA/2/pull_pullx.xvg', comments=['@', '#'])
data6x = np.loadtxt('/data/jychen/MD_projects/SMC/All_Atoms/SteeredMD/2SMD_1protonataing+DNA/3/pull_pullx.xvg', comments=['@', '#'])
data4y = np.loadtxt('/data/jychen/MD_projects/SMC/All_Atoms/SteeredMD/2SMD_1protonataing+DNA/1/pull_pullf.xvg', comments=['@', '#'])
data5y = np.loadtxt('/data/jychen/MD_projects/SMC/All_Atoms/SteeredMD/2SMD_1protonataing+DNA/2/pull_pullf.xvg', comments=['@', '#'])
data6y = np.loadtxt('/data/jychen/MD_projects/SMC/All_Atoms/SteeredMD/2SMD_1protonataing+DNA/3/pull_pullf.xvg', comments=['@', '#'])

# 提取数据每 500 点
data1 = data1x[0:24530:500]
data2 = data1y[0:24530:500]
data3 = data2x[0:24530:500]
data4 = data2y[0:24530:500]
data5 = data3x[0:24530:500]
data6 = data3y[0:24530:500]
data7 = data4x[0:24530:500]
data8 = data4y[0:24530:500]
data9 = data5x[0:24530:500]
data10 = data5y[0:24530:500]
data11 = data6x[0:24530:500]
data12 = data6y[0:24530:500]

# 提取 x 和 y 数据
xdata1 = data1[:, 1]
ydata1 = data2[:, 1]
xdata2 = data3[:, 1]
ydata2 = data4[:, 1]
xdata3 = data5[:, 1]
ydata3 = data6[:, 1]

xdata4 = data7[:, 1]
ydata4 = data8[:, 1]
xdata5 = data9[:, 1]
ydata5 = data10[:, 1]
xdata6 = data11[:, 1]
ydata6 = data12[:, 1]

# 计算第一组数据 (noDNA) 的平均值和误差棒
average_values1 = []
errors1 = []

# 计算第二组数据 (withDNA) 的平均值和误差棒
average_values2 = []
errors2 = []

# 遍历每一帧的数据
for i in range(len(xdata1)):
    # noDNA 组
    avg_noDNA = (ydata1[i] + ydata2[i] + ydata3[i]) / 3
    err_noDNA = np.std([ydata1[i], ydata2[i], ydata3[i]]) / np.sqrt(3)
    average_values1.append(avg_noDNA)
    errors1.append(err_noDNA)

    # withDNA 组
    avg_withDNA = (ydata4[i] + ydata5[i] + ydata6[i]) / 3
    err_withDNA = np.std([ydata4[i], ydata5[i], ydata6[i]]) / np.sqrt(3)
    average_values2.append(avg_withDNA)
    errors2.append(err_withDNA)

# 绘图风格设置
plt.rcParams.update({
    'font.family': 'Arial',  # 设置字体
    'axes.titlesize': 36,  # 标题字体大小
    'axes.labelsize': 36,  # 坐标轴标签字体大小
    'xtick.labelsize': 36,  # X轴刻度字体大小
    'ytick.labelsize': 36,  # Y轴刻度字体大小
    'legend.fontsize': 36,  # 图例字体大小
    'figure.figsize': (8.5, 7),  # 图像大小
    'axes.linewidth': 4,  # 坐标轴线宽
    'xtick.major.size': 10, # x 轴主刻度长度
    'ytick.major.size': 10, # y 轴主刻度长度
    'xtick.minor.size': 5, # x 轴次刻度长度
    'ytick.minor.size': 5, # y 轴主刻度长度
    'xtick.major.width': 4, # x 轴主刻度线宽
    'ytick.major.width': 4, # y 轴主刻度线宽
    'xtick.minor.width': 4, # x 轴次刻度线宽
    'ytick.minor.width': 4, # y 轴主刻度线宽
    'axes.grid': False,  # 关闭背景网格
})

# 绘制堆叠柱状图
fig, ax = plt.subplots(figsize=(8.5, 7))

# 绘制 noDNA 数据及其误差区域
ax.plot(xdata1, average_values1, color='tab:red', linestyle='-', label='US', linewidth=4)
ax.fill_between(
    xdata1,
    np.array(average_values1) - np.array(errors1),
    np.array(average_values1) + np.array(errors1),
    color='tab:red',
    alpha=0.2,
)

# 绘制 withDNA 数据及其误差区域
ax.plot(xdata1, average_values2, color='tab:blue', linestyle='-', label='BS', linewidth=4)
ax.fill_between(
    xdata1,
    np.array(average_values2) - np.array(errors2),
    np.array(average_values2) + np.array(errors2),
    color='tab:blue',
    alpha=0.2,
)

# x 轴刻度朝内
ax.tick_params(axis='x', which='both', direction='in')
# y 轴刻度朝内
ax.tick_params(axis='y', which='both', direction='in')

# 设置y轴的刻度为科学记数法，并只显示数值部分
#formatter = ScalarFormatter()
#formatter.set_scientific(True)
#formatter.set_powerlimits((0, 0))  # 显示科学记数法时，只显示数值部分
#plt.gca().yaxis.set_major_formatter(formatter)

# 加粗刻度字体
#plt.setp(ax.get_xticklabels(), fontweight='bold')
#plt.setp(ax.get_yticklabels(), fontweight='bold')

# 设置刻度
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))  # 横坐标主刻度间隔
#ax.xaxis.set_minor_locator(ticker.MultipleLocator(2.5))  # 横坐标次刻度间隔
ax.yaxis.set_major_locator(ticker.MultipleLocator(200))  # 纵坐标主刻度间隔
#ax.yaxis.set_minor_locator(ticker.MultipleLocator(100))  # 纵坐标次刻度间隔

# 添加网格线
#ax.grid(visible=True, which='major', linestyle='--', linewidth=0.5, alpha=0.7)

# 添加图例
ax.legend(loc='upper right', frameon=False,handlelength=0,labelcolor='linecolor')

# 调整布局
fig.tight_layout(pad=2)

# 设置坐标轴标签和标题
ax.set_xlabel('Distance (nm)')
ax.set_ylabel('Force (kJ/mol/nm)')
#ax.set_title('Force with Distance')

ax.set_xlim(0, 15)
ax.set_ylim(-50, 600)

# 调整图形边距
plt.subplots_adjust(left=0.2, right=0.9, top=0.9, bottom=0.15)  # 固定边距

# 保存高分辨率图片
plt.savefig("Force_with_Distance.png", dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.collections import EventCollection

# read force from the .xvg files
data1 = np.loadtxt('/data/jychen/MD_projects/SMC/All_Atoms/SteeredMD/2SMD_1protonataing/1/pull_pullf.xvg', comments=['@', '#'])
data2 = np.loadtxt('/data/jychen/MD_projects/SMC/All_Atoms/SteeredMD/2SMD_1protonataing/2/pull_pullf.xvg', comments=['@', '#'])
data3 = np.loadtxt('/data/jychen/MD_projects/SMC/All_Atoms/SteeredMD/2SMD_1protonataing/3/pull_pullf.xvg', comments=['@', '#'])

# read force from the .xvg files
data4 = np.loadtxt('/data/jychen/MD_projects/SMC/All_Atoms/SteeredMD/2SMD_1protonataing+DNA/1/pull_pullf.xvg', comments=['@', '#'])
data5 = np.loadtxt('/data/jychen/MD_projects/SMC/All_Atoms/SteeredMD/2SMD_1protonataing+DNA/2/pull_pullf.xvg', comments=['@', '#'])
data6 = np.loadtxt('/data/jychen/MD_projects/SMC/All_Atoms/SteeredMD/2SMD_1protonataing+DNA/3/pull_pullf.xvg', comments=['@', '#'])


#extract data every 500 points
data1xy=data1[0:24530:500]
data2xy=data2[0:24530:500]
data3xy=data3[0:24530:500]
data4xy=data4[0:24530:500]
data5xy=data5[0:24530:500]
data6xy=data6[0:24530:500]



# extract x and y data from the files
xdata1 = data1xy[:, 0]
ydata1 = data1xy[:, 1]
xdata2 = data2xy[:, 0]
ydata2 = data2xy[:, 1]
xdata3 = data3xy[:, 0]
ydata3 = data3xy[:, 1]
xdata4 = data4xy[:, 0]
ydata4 = data4xy[:, 1]
xdata5 = data5xy[:, 0]
ydata5 = data5xy[:, 1]
xdata6 = data6xy[:, 0]
ydata6 = data6xy[:, 1]


# 创建一个新的列表来存储每帧的平均值和误差棒
average_values1 = []
errors1 = []

# 创建一个新的列表来存储第二组数据每帧的平均值和误差棒
average_values2 = []
errors2 = []

# 遍历每一帧的数据
for i in range(len(xdata1)):
    # 对应位置的元素相加并求平均值，计算误差棒，第一组数据
    average = (ydata1[i] + ydata2[i] + ydata3[i]) / 3
    error = np.std([ydata1[i], ydata2[i], ydata3[i]]) / np.sqrt(3)

    # 对应位置的元素相加并求平均值，计算误差棒，第二组数据
    average2 = (ydata4[i] + ydata5[i] + ydata6[i]) / 3
    error2 = np.std([ydata4[i], ydata5[i], ydata6[i]]) / np.sqrt(3)
    
    # 将计算得到的平均值和误差棒存储到新的列表中
    average_values1.append(average)
    errors1.append(error)

    # 将计算得到的平均值和误差棒存储到新的列表中
    average_values2.append(average2)
    errors2.append(error2)

# 绘图风格设置
plt.rcParams.update({
    'font.family': 'Arial',  # 设置字体
    'axes.titlesize': 28,  # 标题字体大小
    'axes.labelsize': 28,  # 坐标轴标签字体大小
    'xtick.labelsize': 28,  # X轴刻度字体大小
    'ytick.labelsize': 28,  # Y轴刻度字体大小
    'legend.fontsize': 28,  # 图例字体大小
    'figure.figsize': (8, 6),  # 图像大小
    'axes.linewidth': 4,  # 坐标轴线宽
    'xtick.major.size': 10, # x 轴主刻度长度
    'ytick.major.size': 10, # y 轴主刻度长度
    'xtick.minor.size': 5, # x 轴次刻度长度
    'ytick.minor.size': 5, # y 轴主刻度长度
    'xtick.major.width': 4, # x 轴主刻度线宽
    'ytick.major.width': 4, # y 轴主刻度线宽
    'xtick.minor.width': 4, # x 轴次刻度线宽
    'ytick.minor.width': 4, # y 轴主刻度线宽
    'legend.fontsize': 28,    # 图例字体大小
    'axes.grid': False,  # 关闭背景网格
})

# 创建图像和子图
fig, ax = plt.subplots(figsize=(8, 6))

# 绘制 noDNA 数据及其误差区域
ax.plot(xdata1, average_values1, color='tab:red', linestyle='-', label='US', linewidth=4)
ax.fill_between(
    xdata1,
    np.array(average_values1) - np.array(errors1),
    np.array(average_values1) + np.array(errors1),
    color='tab:red',
    alpha=0.2,
)

# 绘制 withDNA 数据及其误差区域
ax.plot(xdata1, average_values2, color='tab:blue', linestyle='-', label='BS', linewidth=4)
ax.fill_between(
    xdata1,
    np.array(average_values2) - np.array(errors2),
    np.array(average_values2) + np.array(errors2),
    color='tab:blue',
    alpha=0.2,
)

# x 轴刻度朝内
ax.tick_params(axis='x', which='both', direction='in')
# y 轴刻度朝内
ax.tick_params(axis='y', which='both', direction='in')

# 设置y轴的刻度为科学记数法，并只显示数值部分
#formatter = ScalarFormatter()
#formatter.set_scientific(True)
#formatter.set_powerlimits((0, 0))  # 显示科学记数法时，只显示数值部分
#plt.gca().yaxis.set_major_formatter(formatter)

# 加粗刻度字体
#plt.setp(ax.get_xticklabels(), fontweight='bold')
#plt.setp(ax.get_yticklabels(), fontweight='bold')

# 设置刻度
ax.xaxis.set_major_locator(ticker.MultipleLocator(500))  # 横坐标主刻度间隔
#ax.xaxis.set_minor_locator(ticker.MultipleLocator(2.5))  # 横坐标次刻度间隔
ax.yaxis.set_major_locator(ticker.MultipleLocator(200))  # 纵坐标主刻度间隔
#ax.yaxis.set_minor_locator(ticker.MultipleLocator(100))  # 纵坐标次刻度间隔

# 添加网格线
#ax.grid(visible=True, which='major', linestyle='--', linewidth=0.5, alpha=0.7)

# 添加图例
#ax.legend(loc='upper right', frameon=False,handlelength=0, labelcolor='linecolor')

# 设置坐标轴标签和标题
ax.set_xlabel('Time (ps)', labelpad=10)
ax.set_ylabel('Force (kJ/mol/nm)', labelpad=10)
#ax.set_title('Force with Time')

ax.set_xlim(0, 2500)
ax.set_ylim(-50, 600)

# 调整布局
fig.tight_layout(pad=2)

# 保存图像
plt.savefig("Force_with_Time.png", dpi=300)

# 显示图像
plt.show()
```
