```python
import math
import numpy as np
import MDAnalysis as mda
from itertools import chain
import pandas as pd

def find_continuous_segments(filename):
    """load the data"""
    df = pd.read_csv(filename, delim_whitespace=True, header=None)
    """ create'condition',if A > 0.8 and B > 0.8 is True,otherwise, is False """
    df['condition'] = (df.iloc[:, 1] > 0.8) & (df.iloc[:, 3] > 0.5) & (df.iloc[:, 4] > 0.5)
    """ 通过比较当前行的'condition'与上一行的'condition'是否相同，来标识不同的段落
     'cumsum'用于为每个新的段落分配一个新的组号 """
    df['group'] = (df['condition'] != df['condition'].shift()).cumsum()
    """仅选择满足条件的行，并按照组号聚合，获取每个组的第一个和最后一个'index'"""
    filtered_df = df[df['condition']].groupby('group')[0].agg(['first', 'last'])   
    """将每一对'first'和'last'转换为range格式的字符串"""
    ranges = [f'range({row["first"]}, {row["last"] + 1})' for index, row in filtered_df.iterrows()]   
    return ranges

def calculate_distance(atom1, atom2):
    """calculate the distance between two atoms"""
    return math.sqrt((atom1[0] - atom2[0])**2 + (atom1[1] - atom2[1])**2 + (atom1[2] - atom2[2])**2)

def find_closest_atom(atoms, specific_atom):
    """find the closest atom to a specific atom"""
    min_distance = float('inf')
    closest_atom_index = -1
    for index, atom in enumerate(atoms):
        distance = calculate_distance(atom, specific_atom)
        if distance < min_distance:
            min_distance = distance
            closest_atom_index = index
    return closest_atom_index, min_distance
'''
def cal_dp_index (u,frame_ranges):
    dp_ndx = []
    for ts in frame_ranges:
        u.trajectory[ts]
        atoms = u.atoms.positions
        """select a group of atoms on HEAT binding site and calculate the center of mass"""
        atom_indices = [240,241,242,243,245,246,247]
        group1 = u.atoms[atom_indices]
        center_of_mass = group1.center_of_mass()
        specific_atom = center_of_mass
        """select a group of atoms of ssDNA DB atoms"""
        target_atoms = u.select_atoms("bynum 991:1283 and name DB").positions
        """find the closest atom to the center of mass"""
        closest_atom_index, min_distance = find_closest_atom(target_atoms, specific_atom)
        #print(f"closest_atom_index: {closest_atom_index},min_distance: {min_distance}")
        dp_ndx.append(closest_atom_index)
    return dp_ndx

def cal_slide_speed(dp_index):
    differences = [abs(dp_index[i] - dp_index[i-1]) for i in range(1, len(dp_index))]
    total_difference = sum(differences)
    slide_speed= total_difference/len(differences)
    return slide_speed
'''

def cal_dp_index_with_markers(u, frame_ranges):
    # 将 itertools.chain 转换为列表
    frame_ranges = list(frame_ranges)
    dp_ndx = []
    for i, ts in enumerate(frame_ranges):
        u.trajectory[ts]
        # ... 计算 closest_atom_index ...
        """select a group of atoms on HEAT binding site and calculate the center of mass"""
        atom_indices = [240,241,242,243,245,246,247]
        group1 = u.atoms[atom_indices]
        center_of_mass = group1.center_of_mass()
        specific_atom = center_of_mass
        """select a group of atoms of ssDNA DB atoms"""
        target_atoms = u.select_atoms("bynum 991:1283 and name DB").positions
        """find the closest atom to the center of mass"""
        closest_atom_index, min_distance = find_closest_atom(target_atoms, specific_atom)
        #print(f"closest_atom_index: {closest_atom_index},min_distance: {min_distance}")
        dp_ndx.append((closest_atom_index, i > 0 and frame_ranges[i] == frame_ranges[i-1] + 1))
    return dp_ndx

def cal_slide_speed_with_markers(dp_index):
    differences = []
    for i in range(1, len(dp_index)):
        if dp_index[i][1]:  # Only consider continuous frames
            differences.append(abs(dp_index[i][0] - dp_index[i-1][0]))
    total_difference = sum(differences)
    slide_speed = total_difference / len(differences) if differences else 0
    return slide_speed
```

```python
import MDAnalysis as mda
from itertools import chain
import numpy as np

average_slide_speed = []
total_average_slide_speed = []
total_v_std = []

groups = [0.3,0.5,0.6,0.7]  # Add more group values as needed

for group in groups:
    for i in range(1,51):
        u = mda.Universe(
            f"/hpc2hdd/home/jchen901/data/genesis/GENESIS_Tutorials-2022/Works/SMC_AICG2+/Full_removLig+98bp/021_delPartGo/021_nonspecific_DNAout_0.35/para_{group}/0.35-{i}/run1/pro_dna_test_run.pdb",
            f"/hpc2hdd/home/jchen901/data/genesis/GENESIS_Tutorials-2022/Works/SMC_AICG2+/Full_removLig+98bp/021_delPartGo/021_nonspecific_DNAout_0.35/para_{group}/0.35-{i}/run1/pro_dna_test_run.dcd"
        )
        filename = f"/hpc2hdd/home/jchen901/data/genesis/GENESIS_Tutorials-2022/Works/SMC_AICG2+/Full_removLig+98bp/021_delPartGo/analysis/para_nonspecific{group}/All_knetics/order_parameter/0.35/COLVAR{i}"
        continuous_segments = find_continuous_segments(filename)
        ranges = [eval(r) for r in continuous_segments]
        frame_ranges = chain(*ranges)
        dp_index = cal_dp_index_with_markers(u, frame_ranges)
        slide_speed = cal_slide_speed_with_markers(dp_index)
        average_slide_speed.append(slide_speed)

    ave_v = np.mean(average_slide_speed)
    v_std = np.std(average_slide_speed)
    print(f"Group {group} - average_slide_speed: {ave_v}, v_std: {v_std}")

    total_average_slide_speed.append(ave_v)
    total_v_std.append(v_std)
    average_slide_speed = []  # Reset for the next group
```

```python
import matplotlib.pyplot as plt
import numpy as np

species = ("0.3", "0.5", "0.6", "0.7")
penguin_means = {
    'ave_speed': total_average_slide_speed
}
penguin_std = {
    'ave_speed': total_v_std
}

x = np.arange(len(species))  # the label locations
width = 0.5  # the width of the barsß
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, yerr=penguin_std[attribute], label=attribute, capsize=5)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Frame')
ax.set_title('')
ax.set_xticks(x, species)
ax.legend(loc='upper left', ncols=4)
ax.set_ylim(0, 1)

plt.show()
```
