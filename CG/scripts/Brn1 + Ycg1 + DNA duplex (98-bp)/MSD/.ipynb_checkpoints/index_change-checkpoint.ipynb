# DNA和HEAT相互作用强度为$\sigma=0.6$的情况下,不同$\epsilon$下DNA滑动速度随时间间隔的变化

```python
import math
import numpy as np
import MDAnalysis as mda
from itertools import chain
import pandas as pd
from MDAnalysis.lib.distances import distance_array
import matplotlib.pyplot as plt
import pickle

def calculate_distance(atom1, atom2):
    """calculate the distance between two atoms"""
    return math.sqrt((atom1[0] - atom2[0])**2 + (atom1[1] - atom2[1])**2 + (atom1[2] - atom2[2])**2)

def find_closest_atom(atoms, specific_atom):
    """find the closest atom to a specific atom"""
    min_distance = float('inf')
    closest_atom_index = -1
    for index, atom in enumerate(atoms):
        distance = calculate_distance(atom, specific_atom)
        if distance < min_distance:
            min_distance = distance
            closest_atom_index = index
    return closest_atom_index, min_distance

def cal_dp_index(u):
    dp_ndx = []
    atom_indices = [240, 241, 242, 243, 245, 246, 247]
    group1 = u.atoms[atom_indices]
    target_atoms = u.select_atoms("bynum 991:1283 and name DB")

    for ts in u.trajectory:
        center_of_mass = group1.center_of_mass()
        distances = distance_array(center_of_mass.reshape(1, 3), target_atoms.positions)
        closest_atom_index = np.argmin(distances) # 返回最小值的索引
        closest_atom = target_atoms[closest_atom_index]
        dp_ndx.append(closest_atom.residue.resid)
        #dp_ndx.append(target_atoms.indices[closest_atom_index])
    return dp_ndx

def calculate_rate(indices, step):
    rates = []
    for i in range(0, len(indices) - step, step):
        rate = abs(indices[i + step] - indices[i]) / step
        rates.append(rate)
    return rates

def analyze_rates(dp_index_results, max_step=200):
    rate_means = {}
    rate_stds = {}
    for group, indices_list in dp_index_results.items():
        rate_means[group] = []
        rate_stds[group] = []
        for step in range(1, max_step + 1):
            all_rates = [calculate_rate(indices, step) for indices in indices_list] # 计算某个轨迹在某个step interval下的rate列表, 一共有len(indices-step)个值
            mean_rates = [np.mean(rates) for rates in all_rates] #计算某个轨迹在每个step interval下的rate的平均值
            rate_means[group].append(np.mean(mean_rates)) #计算所有轨迹在某个step interval下的rate的平均值,有50个轨迹就是50个轨迹的平均值
            rate_stds[group].append(np.std(mean_rates) / np.sqrt(len(mean_rates)))  # 计算标准误差
    return rate_means, rate_stds
```

```python
groups = [0.3, 0.35, 0.37, 0.4, 0.45]  # Add more group values as needed
dp_index_results = {}

for group in groups:
    dp_indices = []
    for i in range(1, 51):  # 假设每个 group 有 50 个轨迹
        u = mda.Universe(
            f"/hpc2hdd/home/jchen901/data/genesis/GENESIS_Tutorials-2022/Works/SMC_AICG2+/Full_removLig+98bp/021_delPartGo/021_nonspecific_DNAout_{group}/para_0.6/{group}-{i}/run1/pro_dna_test_run.pdb",
            f"/hpc2hdd/home/jchen901/data/genesis/GENESIS_Tutorials-2022/Works/SMC_AICG2+/Full_removLig+98bp/021_delPartGo/021_nonspecific_DNAout_{group}/para_0.6/{group}-{i}/run1/pro_dna_test_run.dcd"
        )
        dp_index = cal_dp_index(u)
        dp_indices.append(dp_index)
        print("Finished processing group", group, "run", i)
    dp_index_results[group] = dp_indices

#rate_means, rate_stds = analyze_rates(dp_index_results)

# save the results to a pkl file
with open('index1.pkl', 'wb') as f:
    pickle.dump(dp_index_results, f)

# Save the results to a file
#with open('dp_index_results.pkl', 'wb') as f:
#    pickle.dump((rate_means, rate_stds), f)
```

```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
import matplotlib.colors as mcolors

# 更新全局绘图风格
plt.rcParams.update({
    'font.family': 'Arial',  # 设置字体
    'axes.titlesize': 28,  # 标题字体大小
    'axes.labelsize': 28,  # 坐标轴标签字体大小
    'xtick.labelsize': 28,  # X轴刻度字体大小
    'ytick.labelsize': 28,  # Y轴刻度字体大小
    'legend.fontsize': 24,  # 图例字体大小
    'figure.figsize': (8, 6),  # 图像大小
    'axes.linewidth': 3,  # 坐标轴线宽
    'xtick.major.width': 2.4, # x 轴主刻度线宽
    'ytick.major.width': 2.4, # y 轴主刻度线宽
    'xtick.minor.width': 2.1, # x 轴次刻度线宽
    'ytick.minor.width': 2.1, # y 轴次刻度线宽
    'axes.grid': False,  # 关闭背景网格
})

def plot_rates(groups, rate_means, rate_stds):
    cmap = plt.get_cmap('viridis')
    norm = mcolors.Normalize(vmin=min(groups), vmax=max(groups))
    colors = [cmap(norm(group)) for group in groups]  # 使用 coolwarm 颜色映射
    
    linewidths = [2.4] * len(groups)  # 线宽列表

    for i, group in enumerate(groups):
        plt.errorbar(
            range(1, 201), 
            rate_means[group], 
            yerr=rate_stds[group], 
            label=f'$\epsilon_{{LB}} = {group}$',
            color=colors[i % len(colors)],  # 循环颜色
            linewidth=linewidths[i % len(linewidths)],  # 循环线宽
            elinewidth=2.4,  # 误差棒线宽
            capsize=4  # 误差棒端点横线长度
        )
    
    plt.xlim(1, 200)  # 确保 x 轴范围
    plt.ylim(0, 0.3)  # 设置 y 轴范围
    plt.yticks(np.arange(0, 0.4, 0.1))  # 设置 y 轴刻度为 0 到 0.4，步长为 0.1
    plt.xlabel('Step Interval')
    plt.ylabel('DNA Sliding Rate (bp/step)')
    # x 轴刻度朝内
    plt.tick_params(axis='x', which='both', direction='in')
    # y 轴刻度朝内
    plt.tick_params(axis='y', which='both', direction='in')
    plt.title('')
    plt.legend(frameon=False)

    # 添加颜色条
#    sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
#    sm.set_array([])  # 不设置数据
#    cbar = plt.colorbar(sm, ax=plt.gca(), label=r'$\epsilon_\text{HD}$', shrink=0.8)
#    cbar.ax.tick_params(labelsize=20)  # 调整颜色条刻度字体大小

    plt.savefig("DNA_sliding_epsilonHD.png", dpi=300, bbox_inches='tight')
    plt.show()

groups = [0.3, 0.35, 0.37, 0.4, 0.45]  # Add more group values as needed

# Load the results from the file
with open('dp_index_results.pkl', 'rb') as f:
    rate_means, rate_stds = pickle.load(f)

plot_rates(groups, rate_means, rate_stds)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D
import pickle
from matplotlib import cm

# 更新全局绘图风格
plt.rcParams.update({
    'font.family': 'Arial',  # 设置字体
    'axes.titlesize': 28,  # 标题字体大小
    'axes.labelsize': 28,  # 坐标轴标签字体大小
    'xtick.labelsize': 28,  # X轴刻度字体大小
    'ytick.labelsize': 28,  # Y轴刻度字体大小
    'legend.fontsize': 24,  # 图例字体大小
    'figure.figsize': (8, 6),  # 图像大小
    'axes.linewidth': 3,  # 坐标轴线宽
    'xtick.major.size': 10, # x 轴主刻度长度
    'ytick.major.size': 10, # y 轴主刻度长度
    'xtick.minor.size': 5, # x 轴次刻度长度
    'ytick.minor.size': 5, # y 轴主刻度长度
    'xtick.major.width': 2.4, # x 轴主刻度线宽
    'ytick.major.width': 2.4, # y 轴主刻度线宽
    'xtick.minor.width': 2.1, # x 轴次刻度线宽
    'ytick.minor.width': 2.1, # y 轴次刻度线宽
    'axes.grid': False,  # 关闭背景网格
})

def plot_rates(groups, rate_means, rate_stds):
    # 使用 coolwarm 颜色条
    cmap = cm.get_cmap('viridis', len(groups))
    
    linewidths = [4] * len(groups)  # 线宽列表

    # 绘制误差条
    for i, group in enumerate(groups):
        plt.errorbar(
            range(1, 201), 
            rate_means[group], 
            yerr=rate_stds[group], 
            label=f'$\epsilon_{{LB}} = {group}$',
            color=cmap(i),  # 从颜色条中获取颜色
            linewidth=linewidths[i % len(linewidths)],  # 循环线宽
            elinewidth=4,  # 误差棒线宽
            capsize=4  # 误差棒端点横线长度
        )

    # 创建自定义的图例项（不显示线条）
    legend_handles = [
        Line2D([0], [0], color=cmap(i), lw=0, label=rf'$\epsilon_\text{{LB}} = {group}$')
        for i, group in enumerate(groups)
    ]

    # 绘制图例时设置labelcolor为相应的颜色
    plt.xlim(1, 200)  # 确保 x 轴范围
    plt.ylim(0, 0.1)  # 设置 y 轴范围
    plt.yticks(np.arange(0, 0.4, 0.1))  # 设置 y 轴刻度为 0 到 0.4，步长为 0.1
    plt.xlabel('Step Interval')
    plt.ylabel('$V_{DNA}$(bps/interval)')

    # x 轴刻度朝内
    plt.tick_params(axis='x', which='both', direction='in')
    # y 轴刻度朝内
    plt.tick_params(axis='y', which='both', direction='in')
    plt.title('')
    
    # 使用自定义的图例项并且确保文字颜色与曲线颜色一致
    plt.legend(handles=legend_handles, frameon=False, labelcolor=[cmap(i) for i in range(len(groups))])

    plt.savefig("DNA_sliding_epsilonLB.png", dpi=300, bbox_inches='tight')
    plt.show()

groups = [0.3, 0.35, 0.37, 0.4, 0.45]  # Add more group values as needed

# Load the results from the file
with open('dp_index_results.pkl', 'rb') as f:
    rate_means, rate_stds = pickle.load(f)

plot_rates(groups, rate_means, rate_stds)
```
