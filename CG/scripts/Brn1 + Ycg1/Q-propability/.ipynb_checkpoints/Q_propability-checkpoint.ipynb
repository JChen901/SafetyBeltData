```python
#散点图
import matplotlib.pyplot as plt
import numpy as np

# 读取文件数据
data_0_3 = np.loadtxt('0.3_histo/histo_1')
data_0_35 = np.loadtxt('0.35_histo/histo_1')
data_0_37 = np.loadtxt('0.37_histo/histo_1')
data_0_4 = np.loadtxt('0.4_histo/histo_1')
data_0_45 = np.loadtxt('0.45_histo/histo_1')
data_0_5 = np.loadtxt('0.5_histo/histo_1')

# 更新绘图风格
plt.rcParams.update({
    'font.family': 'Arial',  # 设置字体
    'axes.titlesize': 28,  # 标题字体大小
    'axes.labelsize': 28,  # 坐标轴标签字体大小
    'xtick.labelsize': 28,  # X轴刻度字体大小
    'ytick.labelsize': 28,  # Y轴刻度字体大小
    'legend.fontsize': 28,  # 图例字体大小
    'figure.figsize': (8, 6),  # 图像大小
    'axes.linewidth': 3,  # 坐标轴线宽
    'xtick.major.size': 10, # x 轴主刻度长度
    'ytick.major.size': 10, # y 轴主刻度长度
    'xtick.minor.size': 5, # x 轴次刻度长度
    'ytick.minor.size': 5, # y 轴主刻度长度
    'xtick.major.width': 2.4, # x 轴主刻度线宽
    'ytick.major.width': 2.4, # y 轴主刻度线宽
    'xtick.minor.width': 2.1, # x 轴次刻度线宽
    'ytick.minor.width': 2.1, # y 轴主刻度线宽
    'legend.fontsize': 28,    # 图例字体大小
    'axes.grid': False,  # 关闭背景网格
})

# 绘制图像
plt.scatter(data_0_3[:,0],  data_0_3[:,1],  label=r'$\epsilon_\text{LB}$ = 0.3', marker='o', color='#0C4E9B', s= 100)
plt.scatter(data_0_35[:,0], data_0_35[:,1], label=r'$\epsilon_\text{LB}$ = 0.35', marker='s', color='#6B98C4', s= 100)
plt.scatter(data_0_37[:,0],  data_0_37[:,1],  label=r'$\epsilon_\text{LB}$ = 0.37', marker='^', color='#C72228', s= 100)
plt.scatter(data_0_4[:,0], data_0_4[:,1], label=r'$\epsilon_\text{LB}$ = 0.4', marker='*', color='#F5867F', s= 100)
plt.scatter(data_0_45[:,0],  data_0_45[:,1],  label=r'$\epsilon_\text{LB}$ = 0.45', marker='d', color='#F98F34', s= 100)
plt.scatter(data_0_5[:,0],  data_0_5[:,1],  label=r'$\epsilon_\text{LB}$ = 0.5', marker='x', color='#FFBC80', s= 100)

# x 轴刻度朝内
plt.tick_params(axis='x', which='both', direction='in')
# y 轴刻度朝内
plt.tick_params(axis='y', which='both', direction='in')

# 添加图例
plt.legend(frameon=False)

plt.xticks(np.arange(0, 1.1, 0.5),rotation=0)
plt.yticks(np.arange(0, 10, 3),rotation=0)

# 设置坐标轴标签
plt.xlabel('Q')
plt.xlim(0, 1)
plt.ylabel('Probability density')
plt.savefig("Q-Probability-density", dpi=300, bbox_inches='tight')
# 显示图像
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors

# Load data from files
data_0_3 = np.loadtxt('0.3_histo/histo_1')
data_0_35 = np.loadtxt('0.35_histo/histo_1')
data_0_37 = np.loadtxt('0.37_histo/histo_1')
data_0_4 = np.loadtxt('0.4_histo/histo_1')
data_0_45 = np.loadtxt('0.45_histo/histo_1')
data_0_5 = np.loadtxt('0.5_histo/histo_1')

# Combine all data into a list for easier processing
all_data = [data_0_3, data_0_35, data_0_37, data_0_4, data_0_45, data_0_5]
epsilons = [0.3, 0.35, 0.37, 0.4, 0.45, 0.5]  # Corresponding epsilon values

# Update plotting style
plt.rcParams.update({
    'font.family': 'Arial',
    'axes.titlesize': 28,
    'axes.labelsize': 28,
    'xtick.labelsize': 28,
    'ytick.labelsize': 28,
    'legend.fontsize': 28,
    'figure.figsize': (8, 6),
    'axes.linewidth': 3,
    'xtick.major.size': 10,
    'ytick.major.size': 10,
    'xtick.minor.size': 5,
    'ytick.minor.size': 5,
    'xtick.major.width': 2.4,
    'ytick.major.width': 2.4,
    'xtick.minor.width': 2.1,
    'ytick.minor.width': 2.1,
    'legend.fontsize': 28,
    'axes.grid': False,
})

# Normalize epsilon values for color mapping
norm = mcolors.Normalize(vmin=min(epsilons), vmax=max(epsilons))
cmap = plt.get_cmap('viridis')

# Plot scatter with gradient colors
for data, epsilon in zip(all_data, epsilons):
    color = cmap(norm(epsilon))
    plt.scatter(data[:, 0], data[:, 1], label=fr'$\epsilon_\text{{LB}}$ = {epsilon}', 
                color=color, s=100, edgecolor='black')

# Add color bar
#sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
#sm.set_array([])
#cbar = plt.colorbar(sm, ax=plt.gca())
#cbar.set_label(r'$\epsilon_\text{LB}$')

# Axis settings
plt.tick_params(axis='x', which='both', direction='in')
plt.tick_params(axis='y', which='both', direction='in')

# Add legend
plt.legend(frameon=False)

plt.xticks(np.arange(0, 1.1, 0.5), rotation=0)
plt.yticks(np.arange(0, 10, 3), rotation=0)

# Set axis labels
plt.xlabel('Q')
plt.xlim(0, 1)
plt.ylabel('Probability density')
plt.savefig("Q-Probability-density", dpi=300, bbox_inches='tight')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors
import matplotlib.patheffects as PathEffects

# Load data from files
data_0_3 = np.loadtxt('0.3_histo/histo_1')
data_0_35 = np.loadtxt('0.35_histo/histo_1')
data_0_37 = np.loadtxt('0.37_histo/histo_1')
data_0_4 = np.loadtxt('0.4_histo/histo_1')
data_0_45 = np.loadtxt('0.45_histo/histo_1')
data_0_5 = np.loadtxt('0.5_histo/histo_1')

# Combine all data into a list for easier processing
all_data = [data_0_3, data_0_35, data_0_37, data_0_4, data_0_45, data_0_5]
epsilons = [0.3, 0.35, 0.37, 0.4, 0.45, 0.5]  # Corresponding epsilon values

# Update plotting style
plt.rcParams.update({
    'font.family': 'Arial',
    'axes.titlesize': 28,
    'axes.labelsize': 28,
    'xtick.labelsize': 28,
    'ytick.labelsize': 28,
    'legend.fontsize': 28,
    'figure.figsize': (8, 6),
    'axes.linewidth': 3,
    'xtick.major.size': 10,
    'ytick.major.size': 10,
    'xtick.minor.size': 5,
    'ytick.minor.size': 5,
    'xtick.major.width': 2.4,
    'ytick.major.width': 2.4,
    'xtick.minor.width': 2.1,
    'ytick.minor.width': 2.1,
    'legend.fontsize': 28,
    'axes.grid': False,
})

# Normalize epsilon values for color mapping
norm = mcolors.Normalize(vmin=min(epsilons), vmax=max(epsilons))
cmap = plt.get_cmap('viridis')

# Plot lines with gradient colors and black border
lines = []
for data, epsilon in zip(all_data, epsilons):
    color = cmap(norm(epsilon))
    line, = plt.plot(data[:, 0], data[:, 1], label=fr'$\epsilon_\text{{LB}}$ = {epsilon}', 
                     color=color, linewidth=4)  # Increase the line width
    line.set_path_effects([PathEffects.Stroke(linewidth=6, foreground='black'), PathEffects.Normal()])
    lines.append(line)

# Customize legend to show only text with colors matching the lines
legend_labels = [fr'$\epsilon_\text{{LB}}$ = {epsilon}' for epsilon in epsilons]
legend = plt.legend(lines, legend_labels, frameon=False, handlelength=0, handletextpad=0)

for text, epsilon in zip(legend.get_texts(), epsilons):
    color = cmap(norm(epsilon))
    text.set_color(color)

# Axis settings
plt.tick_params(axis='x', which='both', direction='in')
plt.tick_params(axis='y', which='both', direction='in')

plt.xticks(np.arange(0, 1.1, 0.5), rotation=0)
plt.yticks(np.arange(0, 10, 3), rotation=0)

# Set axis labels
plt.xlabel('Q')
plt.xlim(0, 1)
plt.ylabel('Probability density')
plt.savefig("Q-Probability-density", dpi=300, bbox_inches='tight')

# Show plot
plt.show()
```
