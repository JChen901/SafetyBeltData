```python
#散点图
import matplotlib.pyplot as plt
import numpy as np

# 读取文件数据
data_0_0 = np.loadtxt('histo_1')
data_0_1 = np.loadtxt('histo_2')
data_0_2 = np.loadtxt('histo_3')
data_0_3 = np.loadtxt('histo_4')
data_0_4 = np.loadtxt('histo_5')
data_0_5 = np.loadtxt('histo_6')
data_0_6 = np.loadtxt('histo_7')
data_0_7 = np.loadtxt('histo_8')

# 绘制图像
plt.scatter(data_0_0[:,0], data_0_0[:,1], label='histo_0', marker='o', color='r')
plt.scatter(data_0_1[:,0], data_0_1[:,1], label='histo_0.1', marker='s', color='g')
plt.scatter(data_0_2[:,0], data_0_2[:,1], label='histo_0.2', marker='^', color='b')
plt.scatter(data_0_3[:,0], data_0_3[:,1], label='histo_0.3', marker='*', color='c')
plt.scatter(data_0_4[:,0], data_0_4[:,1], label='histo_0.4', marker='x', color='m')
plt.scatter(data_0_5[:,0], data_0_5[:,1], label='histo_0.5', marker='o', color='y')
plt.scatter(data_0_6[:,0], data_0_6[:,1], label='histo_0.6', marker='s', color='k')
plt.scatter(data_0_7[:,0], data_0_7[:,1], label='histo_0.7', marker='^', color='r')

# 添加图例
plt.legend()

# 设置坐标轴标签
plt.xlabel('Fraction of native contacts (Q)')
plt.xlim(0, 1)
plt.ylabel('Probability density')

# 显示图像
plt.show()
```
